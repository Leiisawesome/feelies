```python
from google.colab import drive
import os

drive.mount('/content/drive')

workdir = '/content/drive/MyDrive/feelies_workdir'
if not os.path.exists(workdir):
    os.makedirs(workdir, exist_ok=True)
    print(f'Created working directory: {workdir}')
else:
    print(f'Working directory already exists: {workdir}')
```

```python
import importlib.util
import subprocess
import sys

package_name = 'massive'

try:
    if importlib.util.find_spec(package_name) is None:
        print(f'{package_name} is not installed. Installing...')
        subprocess.check_call([sys.executable, '-m', 'pip', 'install', package_name])
        install_status = 'installed'
    else:
        install_status = 'already installed'
except Exception as exc:
    install_status = f'failed: {exc}'

if importlib.util.find_spec(package_name) is None:
    print(f'Installation status: {install_status}')
    print(f'{package_name} is still unavailable.')
else:
    print(f'Installation status: {install_status}')
    print(f'{package_name} is available now.')
```
